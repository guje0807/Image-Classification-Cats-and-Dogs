```python
import matplotlib.image as image
import matplotlib.pyplot as plt
import numpy as np
import os
import cv2
import pickle 

%matplotlib inline

cat_path = "G:\\Lambton Sem 1\\aml-1413 Introduction to AI\\Assignment-II\\Dataset\\dogscats\\train\\cats"
dog_path = "G:\\Lambton Sem 1\\aml-1413 Introduction to AI\\Assignment-II\\Dataset\\dogscats\\train\\dogs"

number_of_images = 5000  # Taking 5000 images of Dog ans Cats as input
total_imgs = [] #list to store the images as a matrix
```

```python
#Reading the images

image_size = 100   # Setting the image size to 100 * 100
catagories = ['cat','dog']

#Storing Cat images to a list
count = 0
for i in os.listdir(cat_path):
    path = os.path.join(cat_path,i)
    img_arr = cv2.imread(path)
    img_arr = cv2.resize(img_arr,(image_size,image_size))
    total_imgs.append([img_arr,catagories.index('cat')])
    count += 1
    if count==number_of_images:
        break
        
#Storing Dog images to a list 
count = 0
for k in os.listdir(dog_path):
    img_arr = cv2.imread(os.path.join(dog_path,k))
    img_arr = cv2.resize(img_arr,(image_size,image_size))
    total_imgs.append([img_arr,catagories.index('dog')])
    count += 1
    if count==number_of_images:
        break
```
